```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


#data from 7-12 clock
#month_1
mydata1 = pd.read_csv('WeekdayRental1.csv')
#month_2
mydata2 = pd.read_csv('WeekdayRental2.csv')
#duration of the 2 month
mydata3 = pd.read_csv('Duration2.csv')
mydata4 = pd.concat([mydata1,mydata2])
#month_12
mydata5 = pd.read_csv('WeekdayRental12.csv')
mydata6 = pd.read_csv('Duration12.csv')


rate = 0.001
x_train = np.array(mydata4)
print(x_train)
y_train = mydata3['sum']
print(y_train)
x_test  = np.array(mydata5)
print(x_test)


a = np.random.normal()
b = np.random.normal()
c = np.random.normal()
 
def h(x):
    return a*x[0]+b*x[1]+c
 
for i in range(10000):
    sum_a=0
    sum_b=0
    sum_c=0
    for x, y in zip(x_train, y_train):
        sum_a = sum_a + rate*(y-h(x))*x[0]
        sum_b = sum_b + rate*(y-h(x))*x[1]
        sum_c = sum_c + rate*(y-h(x))
    a = a + sum_a
    b = b + sum_b
    c = c + sum_c
    plt.plot([h(xi) for xi in x_test])
 
print(a)
print(b)
print(c)
 
result=[h(xi) for xi in x_train]
print(result)
 
result=[h(xi) for xi in x_test]
print(result)
#real data for comparaton
print(mydata6)

 
plt.show()
```
